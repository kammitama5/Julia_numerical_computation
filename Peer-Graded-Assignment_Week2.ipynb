## My Equation: f(x) : (x^3 )+ 3


```julia
cubed(x::Int64) = (x^3)+3 # takes an integer and returns cube of that integer
```

## Curve that Intersects Y-Axis Once and the X-Axis at two points

```julia
# For loop for values (range) to be plotted
num = 5
x_list = []
y_list = []
for i in -5:num
    push!(x_list, i)
    push!(y_list, (i^3)+3)
end
x_list
# prints out 21-element Array of Any type with 21 elements (-5 through 5)
```

```julia
y_list # prints out y values from y_list that are the cube of what would be in x_list
```

```julia
# Plot with labelled axes
```

```julia

using Plots

plotly() # Choose the Plotly.jl backend for web interactivity
gr() # turn on grid
plot(x_list, y_list,
legend = :bottomright, # sets position of legend
marker = ([:hex], 3), # sets marker for plot
label     = ["y_values" "x_values"],
title      = "Plotting Function x^3+3 over x values",
xlabel   = "x values",
ylabel   = "y values",
)
```
